# Extracción datos de archivo countries.xml

```python
formato = "{0} {1} {2}.\n"

print(formato.format("sujeto", "predicado", "objeto"))
```

```python
countries = open("countries.xml", "r")
classes = set()

for i in countries:
    if "rdf:type" in i:
        classes.add(i)
countries.close()
```

```python
classes
```

```python
clases = {
    "type": "<http://www.w3.org/1999/02/22-rdf-syntax-ns#type>",
    "hasBorderWith": "<http://www.constituteproject.org/ontology/hasBorderWith>",
    "isInGroup": "<http://www.constituteproject.org/ontology/isInGroup>",
    "nameListEN": "<http://www.constituteproject.org/ontology/nameListEN>",
    "nameListES": "<http://www.constituteproject.org/ontology/nameListES>",
    "nameShortEN" : "<http://www.constituteproject.org/ontology/nameShortEN>",
    "nameShortES" : "<http://www.constituteproject.org/ontology/nameShortES>",
    "nameOfficialEN": "<http://www.constituteproject.org/ontology/nameOfficialEN>",
    "nameOfficialES": "<http://www.constituteproject.org/ontology/nameOfficialES>"
         }
```

```python
countriesTtl = open("countries.ttl", "w")

countries = countries = open("countries.xml", "r")
sujeto = ""

for country in countries:
    if "<NamedIndividual" in country:
        sujeto = "<" + country[30:-3] + ">"
        #print(sujeto)
        
    if "<rdf:type" in country:
        objeto = "<" + country[28:-4] + ">"
        countriesTtl.write(formato.format(sujeto, clases["type"],objeto))
    
    if "<hasBorderWith" in country:
        objeto = "<" + country[33:-4] + ">"
        countriesTtl.write(formato.format(sujeto, clases["hasBorderWith"],objeto))
        
    if "<isInGroup" in country:
        objeto = "<" + country[29:-4] + ">"
        countriesTtl.write(formato.format(sujeto, clases["isInGroup"],objeto))
        
    if "<nameListES" in country:
        objeto = country[16:-14]
        if "xml:lang" in objeto:
            objeto = objeto[14:]
        if "rdf:datatype" in objeto:
            objeto = objeto[55:]
        countriesTtl.write(formato.format(sujeto, clases["nameListES"],'"' + objeto + '"'))
        
    if "<nameListEN" in country:
        objeto = country[16:-14]
        if "xml:lang" in objeto:
            objeto = objeto[14:]
        if "rdf:datatype" in objeto:
            objeto = objeto[55:]
        countriesTtl.write(formato.format(sujeto, clases["nameListEN"],'"' + objeto + '"'))
        
    if "<nameOfficialEN" in country:
        objeto = country[20:-18]
        if "xml:lang" in objeto:
            objeto = objeto[14:]
        if "rdf:datatype" in objeto:
            objeto = objeto[55:]
        countriesTtl.write(formato.format(sujeto, clases["nameOfficialEN"],'"' + objeto + '"'))
    
    if "<nameOfficialES" in country:
        objeto = country[20:-18]
        if "xml:lang" in objeto:
            objeto = objeto[14:]
        if "rdf:datatype" in objeto:
            objeto = objeto[55:]
        countriesTtl.write(formato.format(sujeto, clases["nameOfficialES"],'"' + objeto + '"'))
        
    if "<nameShortEN" in country:
        objeto = country[17:-15]
        if "xml:lang" in objeto:
            objeto = objeto[14:]
        if "rdf:datatype" in objeto:
            objeto = objeto[55:]
        countriesTtl.write(formato.format(sujeto, clases["nameShortEN"],'"' + objeto + '"'))
        
    if "<nameShortES" in country:
        objeto = country[17:-15]
        if "xml:lang" in objeto:
            objeto = objeto[14:]
        if "rdf:datatype" in objeto:
            objeto = objeto[55:]
        countriesTtl.write(formato.format(sujeto, clases["nameShortES"],'"' + objeto + '"'))

countries.close()
countriesTtl.close()
```

# Extracción de Eventos (OLD)

```python
clasesEventos = {
    "type": "<http://www.w3.org/1999/02/22-rdf-syntax-ns#type>",
    "event": "<http://www.w3.org/1999/02/22-rdf-syntax-ns#event>",
    "eventId": "<http://www.constituteproject.org/constitution/id>",
    "country": "<http://www.constituteproject.org/constitution/country>",
    "eventType": "<http://www.w3.org/1999/02/22-rdf-syntax-ns#event_type>",
    "eventYear": "<http://www.w3.org/1999/02/22-rdf-syntax-ns#event_year>"
    }
```

```python
import re

about = re.compile('rdf:about="http:.*"')
eventId = re.compile('id="\d*"')
country = re.compile('country=".*"')
```

```python
eventosTtl = open("eventos.ttl", "w")
eventos = open("chronology.xml", "r")

for evento in eventos:
    match = about.search(evento)
    if match != None:
        sujeto = "<" + match.group()[1:-2] + ">"
        eventosTtl.write(formato.format(sujeto, clasesEventos["type"], clasesEventos["event"]))
        
    match = eventId.search(evento)
    if match != None:
        objeto = match.group()[3:]
        eventosTtl.write(formato.format(sujeto,clasesEventos["eventId"] , objeto))
        
    match = country.search(evento)
    if match != None:
        objeto = match.group()[8:]
        eventosTtl.write(formato.format(sujeto, clasesEventos["country"], objeto))
        
    elif "<event_type" in evento:
        objeto = '"' + evento[16:-14] + '"'
        eventosTtl.write(formato.format(sujeto, clasesEventos["eventType"], objeto))
    
    elif "<event_year" in evento:
        objeto = '"' + evento[16:-14] + '"'
        eventosTtl.write(formato.format(sujeto, clasesEventos["eventYear"], objeto))
    
    


eventosTtl.close()
eventos.close()
```

```python
clasesMetadata = {
    "isConstitutionOf" : "<http://www.constituteproject.org/constitution/isConstitutionOf>"
}

resource = re.compile('rdf:resource="http:.*"')
```

```python
metadataTtl = open("metadata.ttl", "w")
metadata = open("metadata.xml", "r")

for metadato in metadata:
    match = about.search(metadato)
    if match != None:
        sujeto = "<" + match.group()[11:-1] + ">"
    
    match = resource.search(metadato)
    if match != None:
        objeto = "<" + match.group()[14:-1] + ">"
        metadataTtl.write(formato.format(sujeto, clasesMetadata["isConstitutionOf"], objeto))
    


metadataTtl.close()
metadata.close()
```

# Extracción eventos (new)

```python
wgcCode = re.compile("WGCODE>\d*<")
countryId = re.compile("country_id>.*<")
metadata = open("metadata.xml", "r")

countriesDict = {}

for metadato in metadata:
        
    match = wgcCode.search(metadato)
    if match != None:
        code = match.group()[7:-1]
    
    match = resource.search(metadato)
    if match != None:
        objeto = "<" + match.group()[14:-1] + ">"
        
        
    match = countryId.search(metadato)
    if match != None:
        country_id = match.group()[11:-1]
        countriesDict[code] = (objeto, country_id)    


metadata.close()
```

```python
dt = pd.read_csv("ccpcce_v1_3.txt", sep="\t")
dt
```

```python
events = dt[dt['evntid'].notnull()]
events = events.astype({'evntid':int})

events = events.sort_values(by = ['evntid'])
events
```

```python
clasesEventos = {
    "type": "<http://www.w3.org/1999/02/22-rdf-syntax-ns#type>",
    "event": "<http://www.w3.org/1999/02/22-rdf-syntax-ns#event>",
    "eventId": "<http://www.constituteproject.org/constitution/id>",
    "country": "<http://www.constituteproject.org/constitution/country>",
    "eventType": "<http://www.w3.org/1999/02/22-rdf-syntax-ns#event_type>",
    "eventYear": "<http://www.w3.org/1999/02/22-rdf-syntax-ns#event_year>",
    "isConstitutionOf" : "<http://www.constituteproject.org/constitution/isConstitutionOf>"
    }
```

```python
newEvents = open("newEvents.ttl", "w")

paisesSinDatos = set()

dictEventos = {}
for indice, row in events.iterrows():
    
    if str(row['cowcode']) in countriesDict:
        objeto = countriesDict[str(row['cowcode'])][0][:-1] + "_" + str(row['year']) + ">"
        dictEventos[row['evntid']] = objeto
        
        newEvents.write(formato.format(objeto, clasesEventos["type"], clasesEventos["event"]))
        newEvents.write(formato.format(objeto, clasesEventos["eventId"], '"' + str(row["evntid"]) + '"'))
        newEvents.write(formato.format(objeto, clasesEventos["country"], '"' + row["country"] + '"'))
        newEvents.write(formato.format(objeto, clasesEventos["eventType"], '"' + row["evnttype"] + '"'))
        newEvents.write(formato.format(objeto, clasesEventos["eventYear"], '"' + str(row["year"]) + '"'))
        newEvents.write(formato.format(objeto, clasesEventos["isConstitutionOf"], countriesDict[str(row['cowcode'])][0]))
        
    else:
        paisesSinDatos.add((row['cowcode'],row['country']))
        
newEvents.close()
```
